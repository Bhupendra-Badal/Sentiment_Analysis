```python
# Load the CSV files into separate DataFrames
df1 = pd.read_excel('youtube_comments12.xlsx')
df2 = pd.read_excel('youtube_comments123.xlsx')
df3 = pd.read_excel('youtube_comments1234.xlsx')
df4 = pd.read_excel('youtube_comments12345.xlsx')

# Concatenate the DataFrames along the rows (axis=0) to merge them vertically
merged_df = pd.concat([df1, df2, df3, df4], axis=0)

# Reset the index of the merged DataFrame to ensure continuous index
merged_df.reset_index(drop=True, inplace=True)

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('updated_merged.csv', index=False)
```
